```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
nir_transmission = pd.read_csv("mktrans_zm_10_10.dat", sep = ' ',  header = None, skipinitialspace=True)
nir_transmission = nir_transmission.apply(pd.to_numeric)
nir_transmission.columns = ['wavelength', 'transmission']


f98 = pd.read_csv("filters/f098m.IR.tab.txt", sep = ' ',  header = None, skipinitialspace=True, skiprows=5)
f98 = f98.drop(0, axis=1)
f98.columns = ["wavelength", "transmission"]
f98.wavelength /= 1e4

f105 = pd.read_csv("filters/f105w.IR.tab.txt", sep = ' ',  header = None, skipinitialspace=True, skiprows=5)
f105 = f105.drop(0, axis=1)
f105.columns = ["wavelength", "transmission"]
f105.wavelength /= 1e4

f110 = pd.read_csv("filters/f110w.IR.tab.txt", sep = ' ',  header = None, skipinitialspace=True, skiprows=5)
f110 = f110.drop(0, axis=1)
f110.columns = ["wavelength", "transmission"]
f110.wavelength /= 1e4

f125 = pd.read_csv("filters/f125w.IR.tab.txt", sep = ' ',  header = None, skipinitialspace=True, skiprows=5)
f125 = f125.drop(0, axis=1)
f125.columns = ["wavelength", "transmission"]
f125.wavelength /= 1e4

f140 = pd.read_csv("filters/f140w.IR.tab.txt", sep = ' ',  header = None, skipinitialspace=True, skiprows=5)
f140 = f140.drop(0, axis=1)
f140.columns = ["wavelength", "transmission"]
f140.wavelength /= 1e4

f160 = pd.read_csv("filters/f160w.IR.tab.txt", sep = ' ',  header = None, skipinitialspace=True, skiprows=5)
f160 = f160.drop(0, axis=1)
f160.columns = ["wavelength", "transmission"]
f160.wavelength /= 1e4


filters = [f98, f105, f110, f125, f140, f160]
labels = ['F908M', 'F105W', 'F110W', 'F125W', 'F140W', 'F160W']
```

```python
fig = plt.figure(figsize=(10, 5), dpi= 300)
ax = fig.gca()
mask = np.where((nir_transmission.wavelength < 2.7) & (nir_transmission.wavelength > .9))
x = nir_transmission.wavelength.to_numpy()[mask]
y = nir_transmission.transmission.to_numpy()[mask]
ax.plot(x, y)

for i, filter in enumerate(filters):
    
    mask = np.where((filter.wavelength <= 2.7) & (filter.wavelength >= .9))
    x = filter.wavelength.to_numpy()[mask]
    y = filter.transmission.to_numpy()[mask]
    ax.plot(x, y, label = labels[i])


ax.set_ylim(0, 1)
ax.set_xlabel("Wavelength ($\mu m$)")
ax.set_ylabel("Transmission")
ax.legend(loc = "upper right")
ax.set_title("Atmospheric Transmission Curve in Near-IR vs. HST IR Filters")
plt.show()
```

### 8)

$F = 1 \times 10^{-17} W m^{-2}$

$\delta = 15 s$

$\sigma = \frac{4F}{\pi \delta^{2}}$

$d_{p} = 25 \mu m$

$s_{p} = \frac{206265 * d_{p}}{f}$

#### Albert:


$D_{A} = .3 m$

$f_{A} = 2.4 m$

$F_{A} = \sigma \frac{\pi D^{2}}{4}$

$E = 100 s_{p}^{2}F_{A} = 100 (\frac{206265 d_{p}}{f_{A}})^{2} \frac{4F}{\pi\delta^{2}}\frac{\pi D_{A}^2}{4}$

```python
F = 1e-7
delta = 15
D = .3
dp = 25e-6
f = 2.4

sigma = 4*F / (np.pi * delta**2)
Fa = sigma * np.pi * D**2 / 4
s = 206265 * dp / f

print("Pixel energy collected for Albert:", np.round(100 * s**2 *Fa, 11), "J")
```

### Bertha:

$D_{B} = 30 m$

$f_{B} = 120 m$

$F_{B} = \sigma \frac{\pi D^{2}}{4}$

$E = 100 s_{p}^{2}F_{B}$


```python
F = 1e-7
delta = 15
D = 30
dp = 25e-6
f = 120

sigma = 4*F / (np.pi * delta**2)
Fb = sigma * np.pi * D**2 / 4
s = 206265 * dp / f

print("Pixel energy collected for Bertha:", np.round(100 * s**2 * Fb, 11), "J")
```

### 9)

For 300 nm:

$\theta = \frac{1.22 \cdot 300 \times 10^{-9}}{2.4 m}$

```python
diffraction_limit = 1.22 * 300e-9 / 2.4 * 206265
print("The diffraction limit for Hubble at 300 nm is:", np.round(diffraction_limit, 3), "arc seconds")
```

At 2 $\mu m$:

$\theta = \frac{1.22 \cdot 2 \times 10^{-6}}{2.4 m}$

```python
diffraction_limit = 1.22 * 2e-6 / 2.4 * 206265
print("The diffraction limit for Hubble at 2 micrometers is:", np.round(diffraction_limit, 3), "arc seconds")
```

For an 8m telescope at 2 $\mu m$:
